```python
class WeeklyPayCalculator:
    def __init__(self, caregivers):
        self.caregivers = caregivers

    def calculate_weekly_pay(self):
        """Calculate weekly pay for all caregivers."""
        return {caregiver.name: caregiver.hours * caregiver.pay_rate for caregiver in self.caregivers}

class PayReport:
    def __init__(self, weekly_pays):
        self.weekly_pays = weekly_pays
        self.weekly_total = sum(weekly_pays.values())
        self.monthly_total = self.weekly_total * 4
        # assuming four weeks per month #

    def generate_text_report(self, filename="pay_report.txt"):
        """Generate a text-based pay report."""
        with open(filename, "w") as file:
            file.write("Weekly Pay Report\n")
            file.write("=" * 50 + "\n")
            for caregiver, pay in self.weekly_pays.items():
                file.write(f"Caregiver: {caregiver}\n")
                file.write(f"Weekly Gross Pay: ${pay}\n")
                file.write("-" * 50 + "\n")
            file.write(f"Weekly Total: ${self.weekly_total}\n")
            file.write(f"Projected Monthly Total: ${self.monthly_total}\n")
        print(f"Pay report saved to {filename}")

# Sample caregivers
caregivers = [
    Caregiver("Linda", "555-1212", "linda@example.com", 20),
    Caregiver("Paul", "555-3434", "paul@example.com", 20),
    Caregiver("Nina", "555-5656", "nina@example.com", 20),
    Caregiver("Oliver", "555-7878", "oliver@example.com", 20),
    Caregiver("Sophia", "555-9090", "sophia@example.com", 20),
    Caregiver("Ethan", "555-2323", "ethan@example.com", 20),
    Caregiver("Maya", "555-4545", "maya@example.com", 20),
    Caregiver("Jack", "555-6767", "jack@example.com", 20),
]

# Adding Hours Worked to Caregivers (Example Data):
caregivers[0].add_hours(40)  # Linda
caregivers[1].add_hours(35)  # Paul
caregivers[2].add_hours(20)  # Nina
caregivers[3].add_hours(30)  # Oliver
caregivers[4].add_hours(25)  # Sophia
caregivers[5].add_hours(45)  # Ethan
caregivers[6].add_hours(38)  # Maya
caregivers[7].add_hours(50)  # Jack



# Calculate Weekly Pay and Generate Reports
pay_calculator = WeeklyPayCalculator(caregivers)
weekly_pays = pay_calculator.calculate_weekly_pay()

# Generate the Report
report = PayReport(weekly_pays)
report.generate_text_report() 

```
